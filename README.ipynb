# Java 学习示例项目

这个仓库用于配合你的 Java 学习计划，包含按阶段划分的示例代码：

- `src/basic`：阶段 1，Java 基础语法与控制台小程序（HelloWorld、计算器、猜数字、TODO CLI）。
- `src/oop`：阶段 2，面向对象与小型学生管理系统示例。
- `src/advanced`：阶段 3，JVM、多线程、I/O 等进阶示例。
- `src/springboot`：阶段 4，Spring Boot 风格的后端示例（仅代码示例，不一定能直接运行）。
- `src/interview`：阶段 5，算法和数据结构题目的 Java 实现示例。
- `src/project`：阶段 6，综合实战项目结构草稿。
